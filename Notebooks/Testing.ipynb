```python
output = []
for file in os.listdir(cwd):
    if not ".json" in file:
        continue
    with open(file) as json_file:  
        data = json.load(json_file)
        for wings in data["wings"]:
                isApis = 0
                if isApisMelliferaMellifera(wings["Ci"], wings["Dv"]):
                    isApis = 1
                output.append(str(wings["Dv"]) + ", " + str(wings["Ci"]) + ", " + str(isApis))
with open("index.csv", "wt") as output_file:
    for row in output:
        output_file.write(row + "\n")
```
